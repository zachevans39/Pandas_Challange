```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
schools_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
schools_complete.head()



```

```python
students =len(schools_complete["Student ID"].unique())
schools = len(schools_complete["school_name"].unique())
budget = school_data['budget'].sum()
math_avg = schools_complete["math_score"].mean()
read_avg = schools_complete["reading_score"].mean()
pass_reading = schools_complete.loc[schools_complete["reading_score"] >= 70] .shape[0]
pass_math = schools_complete.loc[schools_complete["math_score"] >= 70].shape[0] 
overall = schools_complete.loc[(schools_complete["reading_score"]>=70) &(schools_complete["math_score"]>=70)].shape[0]
a=int(pass_reading)
b = int(pass_math)
c= int(students)
d= int(overall)
passing = d/c *100

pass_read = a/c *100
math_pass= b/c *100

```

```python
district_summary = pd.DataFrame ({"Total Schools": [schools],
                                "Total Students": students,
                                "Total Budget": budget ,
                                "Average Math Score":math_avg ,
                                "Average Reading Score":read_avg ,
                                "% Passing Math":math_pass ,
                                "% Passing Reading": pass_read ,
                                "% Overall Passing": passing})

                                  
                            
```

```python
high_schools = schools_complete["school_name"].unique()
az_high = (np.sort(high_schools))
school_type = school_data.set_index(["school_name"])["type"]
schools_students = schools_complete["school_name"].value_counts()
az_schools = schools_students.sort_index()
#school_budget = schools_complete["budget"].unique()
schools_budget = schools_complete.groupby("school_name").mean()["budget"]
students_budget= schools_budget / schools_students
school_math = schools_complete.groupby("school_name").mean()["math_score"]
school_read = schools_complete.groupby("school_name").mean()["reading_score"]
per_school_math = schools_complete[(schools_complete["math_score"] >= 70)]
school_passing_math = per_school_math.groupby(["school_name"]).count()
per_school_reading = schools_complete[(schools_complete["reading_score"] >= 70)]
school_passing_reading = per_school_reading.groupby(["school_name"]).count()
math_ppct = (school_passing_math["math_score"] / az_schools) * 100    
reading_ppct = (school_passing_reading["reading_score"] / az_schools) * 100
#overall_pass = schools_complete[(schools_complete["math_score"] >= 70)] & [(schools_complete["reading_score"] >= 70)]
overall_pass = per_school_math[per_school_math["reading_score"] >= 70]
ovr_ppct = overall_pass.groupby(["school_name"]).count()
passing_pct = (ovr_ppct["math_score"]) / (az_schools) * 100 
                   
```

```python
schools_students
```

```python
school_summary = pd.DataFrame ({"School Name": az_high,
                                "School Type": school_type,
                                "Total Students":az_schools,
                                "Total School Budget": schools_budget,
                                "Per Student Budget" : students_budget,
                                "Average Math Score": school_math,
                                "Average Reading Score": school_read,
                                "% Passing Math": math_ppct,
                                "% Passing Reading":reading_ppct,
                                "% Overall Passing ": passing_pct})
                                
```
